### Implementing BataFrame

```python
df = pd.DataFrame(np.floor(np.random.rand(10,7)*100))
df.to_csv('anim.csv')
# print(df)
# df.describe()
arr=df.to_numpy()
# arr
```

```python
cumsum=np.cumsum(arr[:,:],axis=0)
print(cumsum)
```

### ploting the data

```python
plt.figure(figsize=(15,8))
plt.title('My Data')
years = list(i +2020 for i in range(10))  # â†’ [0, 1, ..., 9]
plt.plot(years, np.mean(cumsum, axis=1))
plt.grid(True)
plt.show()
```
